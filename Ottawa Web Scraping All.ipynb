```python
url="https://www.volunteerottawa.ca/Search_Volunteer_Opportunities.html"
html_content = requests.get(url).text
soup = BeautifulSoup(html_content, "lxml")
```

```python
table = soup.find_all('div', attrs={'class':'single-post-wrapper'})
Pages=[]
pages = soup.find('div', attrs={"class":"PagedIndex"})
page = pages.find_all('a')
for p in page:
    Pages.append(p.get("href"))
print(Pages)
```

## Functions

```python
def get_maininfo(url):
    html_content = requests.get(url).text
    soup = BeautifulSoup(html_content, "lxml")
    table = soup.find_all('div', attrs={'class':'single-post-wrapper'})

    for i in range(len(table)):
        title=table[i].find('div', attrs={'id':'title'})
        Position.append(title.get_text())
        div = table[i].find_all('div')
        org = div[3].get_text().split(":")
        Organization.append(org[1])
        Description.append(div[4].get_text())
        loc = div[5].get_text().split(":")
        Location.append(loc[1])
        deadline = div[6].get_text().split(":")
        Deadline.append(deadline[1])
        webs=table[i].find_all('a', attrs={'class':'more-post'})
        for link in webs:
            Websites.append(link.get('href'))
    return Position, Organization, Description, Location, Deadline, Websites
```

### Extract Info

```python
Position=[]
Organization=[]
Description=[]
Location=[]
Deadline=[]
Websites=[]
```

```python
for link in Pages:
    Position, Organization, Description, Location, Deadline, Websites = get_maininfo(link)
```

```python
print("Position:{}".format(len(Position)))
print("Organization:{}".format(len(Organization)))
print("Description:{}".format(len(Description)))
print("Location:{}".format(len(Location)))
print("Deadline:{}".format(len(Deadline)))
print("Websites:{}".format(len(Websites)))
```

### Output CSV file

```python
df = pd.DataFrame({'Position':Position,'Location':Location, 
                   'Description':Description,
                 'Organization':Organization,
                 'Deadline':Deadline,
                  "Websites": Websites}) 
```

```python
df.to_csv('ottawa_volunteer.csv', index=False, encoding='utf-8_sig')
```
