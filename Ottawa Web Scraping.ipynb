```python
from bs4 import BeautifulSoup
import requests
from selenium import webdriver
import pandas as pd
```

```python
driver = webdriver.Chrome()
```

```python
url="https://www.volunteerottawa.ca/cgi/page.cgi/_memberlog.html?_keyword=&meta_interests%23_set_0=&meta_interests%23_set_1=COVID-19+Virtual+Volunteering&meta_interests%23_set_2=COVID-19+Light+Touch+Volunteering&meta_interests%23_set_3=COVID-19+High+Touch+Volunteering&meta_language%23_set_0=&meta_location%23_set_0=&meta_availability%23_set_0=&meta_volunteer_time_commitment%23_set_0=&meta_org_category%23_set_0=&meta_clientele%23_set_0=&meta_position_target_audience%23_set_0=&type=volunteer_opportunity&section_id=1&log=do_search_form"
html_content = requests.get(url).text
soup = BeautifulSoup(html_content, "lxml")
```

### Extract Info

```python
Position=[]
Organization=[]
Description=[]
Location=[]
Deadline=[]
Websites=[]

table = soup.find_all('div', attrs={'class':'single-post-wrapper'})

for i in range(len(table)):
    title=table[i].find('div', attrs={'id':'title'})
    Position.append(title.get_text())
    div = table[i].find_all('div')
    org = div[3].get_text().split(":")
    Organization.append(org[1])
    Description.append(div[4].get_text())
    loc = div[5].get_text().split(":")
    Location.append(loc[1])
    deadline = div[6].get_text().split(":")
    Deadline.append(deadline[1])
    webs=table[i].find_all('a', attrs={'class':'more-post'})
    for link in webs:
        Websites.append(link.get('href'))
```

### Output CSV file

```python
df = pd.DataFrame({'Position':Position,'Location':Location, 
                   'Description':Description,
                 'Organization':Organization,
                 'Deadline':Deadline,
                  "Websites": Websites}) 
```

```python
df.to_csv('ottawa_volunteer.csv', index=False, encoding='utf-8_sig')
```
