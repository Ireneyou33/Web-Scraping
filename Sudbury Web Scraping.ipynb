```python
def get_pglnks(pglinks):
    for pglink in pglinks:
        res=pglink.find_all('a')
        for a in res:
            if prefix_url+a.get("href") not in pgLnk and len(a.get("href"))>3:
                pgLnk.append(prefix_url+a.get("href"))
    return pgLnk
```

```python
Position=[]
Websites=[]
Organization=[]
Description=[]
Qualifications=[]
Interest=[]
Location=[]
```

```python
for link in pgLnk:
    print(link)
    html_content = requests.get(url).text
    soup = BeautifulSoup(html_content, "lxml")
    table = soup.find_all('div', attrs={'class':'need'})
    for i in range(len(table)):
        titles=table[i].find_all('div', attrs={'class':'title'})
        Position += [title.text for title in titles]
        websites=table[i].find_all('a')
        web=[website.get("href") for website in websites]
        Websites.append(web[0])
        organizations=table[i].find_all('div', attrs={'class':'excerpt'})
        Organization += [org.text for org in organizations]
```

```python
for link in Websites:
    html_content = requests.get(link).text
    soup = BeautifulSoup(html_content, "lxml")
    blocks = soup.find_all('div', attrs={'class':'section-inner'})
    for i in range(len(blocks)):
        content=blocks[i].text
        if re.match(r"^\s+Description", content):
            Description.append(content)
        elif re.match(r"^\s+Addition", content):
            Qualifications.append(content)
        elif re.match(r"^\s+Intere", content):
            Tag=[]
            tags = blocks[i].find_all('div', attrs={'class':'icon-wrap interest-icon-wrap A4D555'})
            Tag+=[tag.get("title") for tag in tags]     
            Interest.append(Tag)
        elif re.match(r"^\s+Locati", content):
            Location.append(content)
#     assert len(Qualifications) == len(Description)
#     assert len(Interest) == len(Description)
    while len(Location) < len(Description):
        Location.append("nan")
    assert len(Location) == len(Description)
```
